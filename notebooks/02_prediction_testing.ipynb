```python
import requests
import joblib
import pandas as pd

model = joblib.load('C:\\Users\\Devarsh\\OneDrive\\Desktop\\smart-traffic-predictor\\model\\traffic_congestion_model.pkl')

print("Model loaded successfully!")
```

```python
def predict_congestion(temp, rain, snow, clouds, hour, day_of_week, month, is_weekend):
    
    input_data = pd.DataFrame([{
        'temp': temp,
        'rain_1h': rain,
        'snow_1h': snow,
        'clouds_all': clouds,
        'hour': hour,
        'day_of_week': day_of_week,
        'month': month,
        'is_weekend': is_weekend
    }])
    
    prediction = model.predict(input_data)
    
    return prediction[0]
```

```python
def get_ahmedabad_weather():
    
    url = f"https://api.openweathermap.org/data/2.5/weather?q=Ahmedabad,IN&appid={API_KEY}&units=metric"
    
    response = requests.get(url)
    data = response.json()
    
    if response.status_code != 200:
        raise Exception(f"API Error: {data}")
    
    temp = data['main']['temp']
    rain = data.get('rain', {}).get('1h', 0)
    snow = data.get('snow', {}).get('1h', 0)
    clouds = data['clouds']['all']
    
    return temp, rain, snow, clouds
```

```python
get_ahmedabad_weather()
```

```python
from datetime import datetime

def predict_live_ahmedabad():
    
    # Get live weather
    temp, rain, snow, clouds = get_ahmedabad_weather()
    
    # Get current time info
    now = datetime.now()
    hour = now.hour
    day_of_week = now.weekday()
    month = now.month
    is_weekend = 1 if day_of_week in [5,6] else 0
    
    # Predict using model
    prediction = predict_congestion(
        temp,
        rain,
        snow,
        clouds,
        hour,
        day_of_week,
        month,
        is_weekend
    )
    
    return prediction
```

```python
predict_live_ahmedabad()
```

```python
import plotly.express as px
```

```python
def show_ahmedabad_map():
    
    prediction = predict_live_ahmedabad()
    
    # Ahmedabad coordinates
    lat = 23.0225
    lon = 72.5714
    
    color_map = {
        "Low": "green",
        "Medium": "orange",
        "High": "red"
    }
    
    fig = px.scatter_mapbox(
        lat=[lat],
        lon=[lon],
        color=[prediction],
        color_discrete_map=color_map,
        zoom=10,
        height=500
    )
    
    fig.update_layout(
        mapbox_style="open-street-map",
        title=f"Ahmedabad Traffic Status: {prediction}"
    )
    
    fig.show()
```

```python
show_ahmedabad_map()
```
